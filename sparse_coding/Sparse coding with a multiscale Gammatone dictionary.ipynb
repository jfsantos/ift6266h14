```python
import matplotlib.pyplot as plt
import scipy as sc
import numpy as np

# Load signal used to test stuff
from scipy.io import wavfile

fs, data = wavfile.read("/home/jfsantos/Documents/speech_training.wav")
data = data/2.0**15

# Get just 2 minutes of speech to save a little bit of time
training_data = data[0:fs*120]

training_data = training_data.reshape(training_data.shape[0]/1024,1024)
```

```python
# Example of a speech frame
plt.plot(training_data[10])
Audio(data=training_data[10],rate=fs)
```

```python
from sparse_coding_gammatone import gammatone_matrix, erb_space

resolution = 8000
step = 64
b = 1.019
n_channels = 50

# Compute a wavelet dictionary
D_multi = np.r_[tuple(gammatone_matrix(b, fc, resolution, step)
                       for fc in erb_space(150, 8000, n_channels))]
```

```python
import itertools
# Example of an entry from the dictionary
fig, axes = plt.subplots(6,6)
fig.set_size_inches(16,12)
for i, j in itertools.product(range(1,7), range(6)):
    axes[i-1][j].plot(D_multi[6*(i-1)+j])
```

```python
from scikits.talkbox import segment_axis

# Get sample speech segment to reconstruct
test_data = segment_axis(data[fs*200:fs*210], 8000, overlap=0)
print test_data.shape

# Reconstruct it frame-by-frame
from sklearn.decomposition import SparseCoder

coder = SparseCoder(dictionary = D_multi, transform_n_nonzero_coefs=None, transform_alpha=10, transform_algorithm="omp")

result = coder.transform(test_data)
```

```python
orig = data[fs*200:fs*210]
out = np.zeros(orig.shape)

for n in range(result.shape[0]):
    out[n*1024:(n+1)*1024] = np.sum(D_multi.T*result[n],axis=1)
```

```python
fig, axes = plt.subplots(3)
fig.set_size_inches(10,8)

axes[0].plot(orig)
axes[1].plot(out, 'g')
axes[2].plot((out-orig)**2, 'r')
```

```python
wavfile.write('reconst_gammatone.wav', 16000, np.asarray(out, dtype=np.float32))
#wavfile.write('orig.wav', 16000, np.asarray(orig, dtype=np.float32))
```

```python
# Test different sparsity values and correspondent outputs

K = [1, 2, 5, 10, 20, 50, 100, 150]
out_k = np.zeros((len(orig),len(K)))

for k in range(len(K)):
    coder = SparseCoder(dictionary = D_multi, transform_n_nonzero_coefs=K[k], transform_alpha=None, transform_algorithm="omp")
    result = coder.transform(test_data)
    for n in range(result.shape[0]):
        out_k[n*1024:(n+1)*1024,k] = np.sum(D_multi.T*result[n],axis=1)

```

```python
fig, axes = plt.subplots(len(K)+1)
fig.set_size_inches(10,15)

axes[0].plot(orig)

for k in range(len(K)):
    axes[k+1].plot(out_k[:,k],'g')
```

```python
# Use alpha as constraint instead of K

alpha = [10, 5, 2.5, 1, 0.5, 0.1]

out_alpha = np.zeros((len(orig),len(alpha)))

for a in range(len(alpha)):
    coder = SparseCoder(dictionary = D_multi, transform_n_nonzero_coefs=None, transform_alpha=alpha[a], transform_algorithm="omp")
    result = coder.transform(test_data)
    print(sum(result.ravel()!=0))
    for n in range(result.shape[0]):
        out_alpha[n*1024:(n+1)*1024,a] = np.sum(D_multi.T*result[n],axis=1)

fig, axes = plt.subplots(len(alpha)+1)
fig.set_size_inches(10,15)

axes[0].plot(orig)

for a in range(len(alpha)):
    axes[a+1].plot(out_alpha[:,a],'g')
```
